```python
def process_df_by_dataset(df, dataset_name):
    grouped_df = df[df['dataset'] == dataset_name].groupby('epoch')[['eval_loss', 'loss']].mean()
    
    loss = grouped_df['loss'].to_numpy()
    val_loss = grouped_df['eval_loss'].to_numpy()
    
    # assert len(loss) == len(df['epoch'].unique())
    # assert len(val_loss) == len(df['epoch'].unique())
    
    return loss, val_loss
```

### val set | finetuning (1e-4) vs. single-train (1e-4)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
FINETUNE_PATH = 'rs/finetune_val/training_hist.csv'
SINGLETRAIN_PATH = 'rs/single_val/training_hist.csv'

ft_training_hist = pd.read_csv(FINETUNE_PATH)
st_training_hist = pd.read_csv(SINGLETRAIN_PATH)

datasets = st_training_hist['dataset'].unique()
```

```python
for i in range(0, len(datasets), 4):
    
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25, 5))
    
    ###
    try:
        ft_loss, ft_val_loss = process_df_by_dataset(ft_training_hist, datasets[i])
        st_loss, st_val_loss = process_df_by_dataset(st_training_hist, datasets[i])
        
        ax1.set_title(datasets[i])
        ax1.plot(ft_loss, label='ft | loss')
        ax1.plot(ft_val_loss, label='ft | val_loss')
        ax1.plot(st_loss, label='st | loss')
        ax1.plot(st_val_loss, label='st | val_loss')
    except:
        continue    
    
    ###
    try:
        ft_loss, ft_val_loss = process_df_by_dataset(ft_training_hist, datasets[i+1])
        st_loss, st_val_loss = process_df_by_dataset(st_training_hist, datasets[i+1])
        
        ax2.set_title(datasets[i+1])
        ax2.plot(ft_loss, label='ft | loss')
        ax2.plot(ft_val_loss, label='ft | val_loss')
        ax2.plot(st_loss, label='st | loss')
        ax2.plot(st_val_loss, label='st | val_loss')
    except:
        continue
    
    ###
    try:
        ft_loss, ft_val_loss = process_df_by_dataset(ft_training_hist, datasets[i+2])
        st_loss, st_val_loss = process_df_by_dataset(st_training_hist, datasets[i+2])
        
        ax3.set_title(datasets[i+2])
        ax3.plot(ft_loss, label='ft | loss')
        ax3.plot(ft_val_loss, label='ft | val_loss')
        ax3.plot(st_loss, label='st | loss')
        ax3.plot(st_val_loss, label='st | val_loss')
    except:
        continue
    
    ###
    try:
        ft_loss, ft_val_loss = process_df_by_dataset(ft_training_hist, datasets[i+3])
        st_loss, st_val_loss = process_df_by_dataset(st_training_hist, datasets[i+3])
        
        ax4.set_title(datasets[i+3])
        ax4.plot(ft_loss, label='ft | loss')
        ax4.plot(ft_val_loss, label='ft | val_loss')
        ax4.plot(st_loss, label='st | loss')
        ax4.plot(st_val_loss, label='st | val_loss')
    except:
        continue
   
    plt.tight_layout()
    
    plt.legend()
    plt.show()
```
